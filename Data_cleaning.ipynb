```python
#Data Clean-up
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

#df = pd.read_csv('Resources/all.csv')
df = pd.read_csv('all.csv')
df = pd.DataFrame(df)
df.columns
```

```python
#Drop the columns filled with a ton of NAs
df = df.drop(['Urban','Scheduled.Caste.population','Percentage...SC.to.total','Scheduled.Tribe.population',
              'Percentage.to.total.population..ST.','Males..Literate','Females..Literate','Males..Literatacy.Rate',
              'Females..Literacy.Rate','X0...4.years','X5...14.years','X15...59.years','X60.years.and.above..Incl..A.N.S..',
              'Total.workers','Main.workers','Marginal.workers','Non.workers','SC.1.Name','SC.1.Population','SC.2.Name',
              'SC.2.Population','SC.3.Name','SC.3.Population','ST.1.Name','ST.1.Population','ST.2.Name','ST.2.Population',
              'ST.3.Name','ST.3.Population','Imp.Town.1.Name','Imp.Town.1.Population','Imp.Town.2.Name','Imp.Town.2.Population',
              'Imp.Town.3.Name','Imp.Town.3.Population','Total.Inhabited.Villages','Primary.Health.Sub.Centre',
              'College','Electricity..domestic.','Electricity..Agriculture.'],axis=1)
df.head()
#df.isna().sum()
```

```python
#Change '-' to NaN so that we can drop those 
df = df.replace('-',np.nan)
df.head()
```

```python
#Rename Columns
df = df.rename(columns={'Growth..1991...2001.':'Growth Rate',
                        'Number.of.households':'Number of Households',
                        'Household.size..per.household.':'Household Size per Household',
                        'Sex.ratio..females.per.1000.males.':'Females per 1000 Males',
                        'Sex.ratio..0.6.years.':'Sex Ratio,0-6 Year Olds',
                        'Persons..literate':'Literate Persons',
                        'Persons..literacy.rate':'Literacy Rate',
                        'Total.Educated': 'Total Educated',
                        'Data.without.level':'Total Uneducated',
                        'Below.Primary':'Below Primary',
                        'Graduate.and.Above':'Graduate Degree or Above',
                        'Religeon.1.Name':'1st Religion Name',
                        'Religeon.1.Population':'1st Religion Population',
                        'Religeon.2.Name':'2nd Religion Name',
                        'Religeon.2.Population':'2nd Religion Population',
                        'Religeon.3.Name':'3rd Religion Name',
                        'Religeon.3.Population':'3rd Religion Population',
                        'Drinking.water.facilities':'Drinking Water Facilities',
                        'Safe.Drinking.water':'Safe Drinking Water Facilities',
                        'Electricity..Power.Supply.':'Electric Supply',
                        'Primary.school':'Primary School',
                        'Middle.schools':'Middle Schools',
                        'Secondary.Sr.Secondary.schools':'Secondary Schools',
                        'Medical.facility':'Medical Facilities',
                        'Primary.Health.Centre':'Primary Health Centers',
                        'Post..telegraph.and.telephone.facility':'Communication Facilities',
                        'Bus.services':'Bus Services',
                        'Paved.approach.road':'Paved Approach Road',
                        'Mud.approach.road':'Mud Approach Road',
                        'Permanent.House':'Permanent Housing Rate',
                        'Semi.permanent.House':'Semi-Permanent Housing Rate',
                        'Temporary.House':'Temporary Housing Rate'})
df.head()
```

```python
#Remove the leading numbers from religion names
df['1st Religion Name'] = df['1st Religion Name'].str.replace('1.','')
df['2nd Religion Name'] = df['2nd Religion Name'].str.replace('2.','')
df['3rd Religion Name'] = df['3rd Religion Name'].str.replace('3.','')
df.head()
```

```python
df.dtypes
```

```python
#Create some new columns for the future analysis
df['Rate of Clean Water Facilities'] = df.loc[:,'Safe Drinking Water Facilities']/df.loc[:,'Drinking Water Facilities']
df['Communication Facilities per Capita'] = df.loc[:,'Communication Facilities']/df.loc[:,'Persons']
df['Drinking Facilities per Capita'] = df.loc[:,'Drinking Water Facilities']/df.loc[:,'Persons']
df['Medical Facilities'] = pd.to_numeric(df['Medical Facilities'])
df['Medical Facilities per Capita'] = df.loc[:,'Medical Facilities']/df.loc[:,'Persons']
df['Permanent Housing Rate'] = df.loc[:,'Permanent Housing Rate']/100
df['Semi-Permanent Housing Rate'] = df.loc[:,'Semi-Permanent Housing Rate']/100
df['Temporary Housing Rate'] = df.loc[:,'Temporary Housing Rate']/100
df['Permanent Housing'] = round(df.loc[:,'Permanent Housing Rate']*df.loc[:,'Persons'],0)
df['Semi-Permanent Housing'] = round(df.loc[:,'Semi-Permanent Housing Rate']*df.loc[:,'Persons'],0)
df['Temporary Housing'] = round(df.loc[:,'Temporary Housing Rate']*df.loc[:,'Persons'],0)
df['Total Housing'] = df.loc[:,'Permanent Housing'] + df.loc[:,'Semi-Permanent Housing'] + df.loc[:,'Temporary Housing']
df.head()
```

```python
#Drop the NAs
df = df.dropna()
```

```python
#Export as 'cleaned.csv'
df.to_csv("cleaned.csv", index=False, header=True)
```
